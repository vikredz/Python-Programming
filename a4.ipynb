```python
import os
from urllib.request import urlretrieve

def download_text(url):
    local_fname = os.path.basename(url)
    if not os.path.exists(local_fname):
        urlretrieve(url, local_fname)
    return local_fname
```

<b1><u><b>Question 2:  Find Non-ASCII Characters  </b></u></b1>

```python
def non_ascii(text):
#empty list to store non_ascii charecters   
    list2=[]
    
    for line in text:
#using not in operator to avoid repetition of chars.      
        for c in line:
            if ord(c)>=128 and c not in list2:
                list2.append(c)
                
    return list2
```

<b1><u><b>Question 3:  Convert Quotes  </b></u></b1>

```python
def convert_quotes(text):
# x is an empty srting to store text with converted Quotes
    x=""
#mapping 
    dict1={"‘":"'",
           "’":"'",
           "“":'"',
           "”":'"',
           "«":'"',
           "»":'"'}
#using comprehension    

# adding converted text into emmpty string x
    list3=[x.join([dict1.get(c, c) for c in line]) for line in text]
    converted='\n'.join(list3)
    return converted

    
# https://nbviewer.org/url/faculty.cs.niu.edu/~dakoop/cs503-2023sp/notebooks/lecture08.ipynb "Comprehensions"
```

<b1><u><b>Question 4:  Convert Underscores (Italics) </b></u></b1>

```python
import re

def convert_italics(x):
#non-whitespace_word boundary
    a=re.compile(r'(\S)_(?=\b)')

#boundary_non-whitespace character
    b=re.compile(r'(?<=\b)_(\S)')

# Replacing with </i> and <i> 
    x = a.sub(r'</i>',x)
    x = b.sub(r'<i>',x)

    return x

# Reference:https://www.educative.io/answers/what-is-the-resub-function-in-python
# https://pynative.com/python-regex-compile/
```

<b1><u><b>Question 5:  Count Number of Headers and Roman Numerals </b></u></b1>

```python
import re

def count(text):
    header=0
    roman=0
    
    for line in text:
# checks for uppercase and  if doesn't match for Roman numerals increments header count 
        if line.isupper() and not re.match(r'^(?=[MDCLXVI])M*(C[MD]|D?C{0,3})(X[CL]|L?X{0,3})(I[XV]|V?I{0,3})$',line):
            header=header+1
            
#if matches for Roman numerals,increments roman count            
        elif re.match(r'^(?=[MDCLXVI])M*(C[MD]|D?C{0,3})(X[CL]|L?X{0,3})(I[XV]|V?I{0,3})$',line):
            roman=roman+1
    
    return header,roman


# Reference:https://www.w3schools.com/python/ref_string_isupper.asp
#https://www.educative.io/answers/what-is-the-rematch-method-in-python
```

```python
header,roman=count(l2)
print("Number of headers:",header)
print("Number of Roman numerals:", roman)
```
